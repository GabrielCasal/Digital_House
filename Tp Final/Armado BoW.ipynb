# BoW

```python
from sklearn.feature_extraction import text 
stop_words = text.ENGLISH_STOP_WORDS
```

```python
count_vect = CountVectorizer(ngram_range = (1,2), max_df = 0.2, min_df = 0.00001,stop_words=stop_words, lowercase=True)
```

```python
x_train = count_vect.fit_transform(train['text'])
```

```python
x_test = count_vect.transform(test['text'])
```

```python
# Get feature names
feature_names = count_vect.get_feature_names()

# View feature names
feature_names
```

```python
y_train = train['Label']
y_test = test['Label']
```

```python
y_train.shape
```

```python
from scipy import sparse

sparse.save_npz("./Data/X_train_BoW.npz", x_train)
sparse.save_npz("./Data/X_test_BoW.npz", x_test)
y_train.to_csv("./Data/y_train.csv", index=False )
y_test.to_csv("./Data/y_test.csv", index=False )
```
