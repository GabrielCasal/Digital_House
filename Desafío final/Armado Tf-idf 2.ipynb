# Tf-idf

```python
from sklearn.feature_extraction import text 
from sklearn.feature_extraction.text import TfidfVectorizer
stop_words = text.ENGLISH_STOP_WORDS
```

```python
tfidf_vect = TfidfVectorizer(ngram_range = (1,3),max_df = 0.2, min_df = 0.00001,stop_words=stop_words, lowercase=True)
```

```python
x_train = tfidf_vect.fit_transform(train['text'])
```

```python
x_test = tfidf_vect.transform(test['text'])
```

```python
# Get feature names
feature_names = tfidf_vect.get_feature_names()

# View feature names
feature_names
```

```python
from scipy import sparse

sparse.save_npz("./Data/X_train_Tf-idf_2.npz", x_train)
sparse.save_npz("./Data/X_test_Tf-idf_2.npz", x_test)
```
