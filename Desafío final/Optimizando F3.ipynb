```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, precision_score, roc_curve, roc_auc_score, auc, f1_score
from sklearn.externals import joblib
```

```python
X_train = pd.read_csv('./Data/X_train_ensamble_final.csv')
X_test = pd.read_csv('./Data/X_test_ensamble_final.csv')
y_train = pd.read_csv("./Data/y_train.csv", header=None)
y_test = pd.read_csv("./Data/y_test.csv", header=None)
```

```python
y_train = y_train[0]
y_test = y_test[0]
```

```python
modelo = joblib.load('./Modelos/Ensamble/Ensamble_BoW-Otros_x2.sav')
```

```python
y_probs = modelo.predict_proba(X_train)
```

```python
y_probs = y_probs[:,1]
```

```python
y_probs
```

```python
def fn_score(recall, precision, n):
    fn = (n ** 2 + 1) * recall * precision / ((n **2) * precision + recall)
    return fn
```

```python
import seaborn as sns
for j in range(9):
    corte = 0 + 0.1 * j
    y_pred = (y_probs > corte).astype(int)
    print ("Treshold: ", corte) 
    print ("Accuracy: ",accuracy_score(y_train, y_pred))
    print ("Recall: ",recall_score(y_train, y_pred))
    print ("Precision: ",precision_score(y_train, y_pred))
    print ("F1: ",fn_score(recall_score(y_train, y_pred), precision_score(y_train, y_pred), 1))
    print ("F3: ",fn_score(recall_score(y_train, y_pred), precision_score(y_train, y_pred), 3),"\n")
```

```python
import seaborn as sns
for j in range(9):
    corte = 0 + 0.01 * j
    y_pred = (y_probs > corte).astype(int)
    print ("Treshold: ", corte) 
    print ("Accuracy: ",accuracy_score(y_train, y_pred))
    print ("Recall: ",recall_score(y_train, y_pred))
    print ("Precision: ",precision_score(y_train, y_pred))
    print ("F1: ",fn_score(recall_score(y_train, y_pred), precision_score(y_train, y_pred), 1))
    print ("F3: ",fn_score(recall_score(y_train, y_pred), precision_score(y_train, y_pred), 3),"\n")
```

```python
y_probs_test = modelo.predict_proba(X_test)
```

```python
y_probs_test = y_probs_test[:,1]
```

```python
y_pred_test = (y_probs_test > 0.05).astype(int)
```

```python
y_pred_test
```

```python
print ("Accuracy: ",accuracy_score(y_test, y_pred_test))
print ("Recall: ",recall_score(y_test, y_pred_test))
print ("Precision: ",precision_score(y_test, y_pred_test))
print ("F1: ",fn_score(recall_score(y_test, y_pred_test), precision_score(y_test, y_pred_test), 1))
print ("F3: ",fn_score(recall_score(y_test, y_pred_test), precision_score(y_test, y_pred_test), 3),"\n")
```

```python
mat = confusion_matrix(y_test, y_pred_test)
sns.set_style('white')
sns.heatmap(mat.T, annot=True, fmt='d')
plt.xlabel('Etiquetas verdaderas')
plt.ylabel('Etiquetas predichas');
```
