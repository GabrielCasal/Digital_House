# BoW

```python
from sklearn.feature_extraction import text 
stop_words = text.ENGLISH_STOP_WORDS
```

```python
count_vect = CountVectorizer(ngram_range = (1,2), max_df = 0.2, min_df = 0.00005,stop_words=stop_words, lowercase=True)
```

```python
x_train = count_vect.fit_transform(train['text'])
```

```python
x_test = count_vect.transform(test['text'])
```

```python
y_train = train['Label']
y_test = test['Label']
```

```python
# Get feature names
feature_names = count_vect.get_feature_names()

# View feature names
feature_names
```

```python
from scipy import sparse

sparse.save_npz("./Data/X_train_BoW_2.npz", x_train)
sparse.save_npz("./Data/X_test_BoW_2.npz", x_test)
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix, recall_score, accuracy_score, precision_score, roc_curve, roc_auc_score, auc
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
modelo_NB = GridSearchCV(BernoulliNB(),param_grid={},scoring='roc_auc', verbose= 10, n_jobs=-1, cv=3)
```

```python
modelo_NB.fit(x_train, y_train)
y_probs = modelo_NB.predict_proba(x_test)[:,1]
fpr_nb,tpr_nb,thr_nb = roc_curve(y_test, y_probs)
auc(fpr_nb, tpr_nb)
```
